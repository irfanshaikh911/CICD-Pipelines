```python
# Wine Quality Analysis using Gaussian Distribution

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix

# Uncomment this to load real dataset
# df = pd.read_csv('winequality-red.csv')  # or 'winequality-white.csv'

# Generate synthetic wine data
np.random.seed(42)
n_samples = 1000

# Features: alcohol, sulphates, citric acid, pH
alcohol = np.random.normal(loc=10, scale=1.5, size=n_samples)
sulphates = np.random.normal(loc=0.6, scale=0.15, size=n_samples)
citric_acid = np.random.normal(loc=0.3, scale=0.1, size=n_samples)
pH = np.random.normal(loc=3.2, scale=0.2, size=n_samples)

# Quality depends on alcohol and citric acid (synthetically)
quality = (alcohol + citric_acid * 10 + np.random.normal(0, 1, n_samples)).round()
quality = np.clip(quality, 3, 8)  # Bound quality between 3 and 8

# Create dataframe
df = pd.DataFrame({
    'alcohol': alcohol,
    'sulphates': sulphates,
    'citric_acid': citric_acid,
    'pH': pH,
    'quality': quality.astype(int)
})

# Plot feature distributions (Gaussian assumption check)
plt.figure(figsize=(12, 6))
for i, col in enumerate(['alcohol', 'sulphates', 'citric_acid', 'pH']):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Prepare data
X = df[['alcohol', 'sulphates', 'citric_acid', 'pH']]
y = df['quality']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Gaussian Naive Bayes model (statistical model using Gaussian dist.)
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Evaluate
print("Classification Report:\n")
print(classification_report(y_test, y_pred))

# Confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_mat, annot=True, cmap="Blues", fmt='d',
            xticklabels=sorted(df['quality'].unique()),
            yticklabels=sorted(df['quality'].unique()))
plt.title("Confusion Matrix - Wine Quality Classification")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```
